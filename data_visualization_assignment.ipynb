### In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

Import required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris_dataset = load_iris()
```

```python
iris_df = pd.DataFrame(iris_dataset.data,columns=iris_dataset.feature_names)
```

```python
iris_df.shape
```

```python
iris_df.head()
```

```python
target = iris_dataset.target
```

### Use PCA to decompose the dataset to 3 dimensions to get a better understanding of the interaction

```python
iris_dataset_reduced = PCA(n_components=3).fit_transform(iris_df)
```

```python
iris_dataset_reduced
```

```python
fig= plt.figure(1,figsize= (8,6))
ax = Axes3D(fig,elev=-150, azim=110)
ax.scatter(iris_dataset_reduced[:,0],iris_dataset_reduced[:,1],iris_dataset_reduced[:,2],
           c=target,cmap=plt.cm.Set2,edgecolor='k', s=40)
ax.set_title("Iris dataset decomposed - First 3 PCA directions")
ax.set_xlabel("1st eigenvector")
ax.set_ylabel("2nd eigenvector")
ax.set_zlabel("3rd eigenvector")
```
